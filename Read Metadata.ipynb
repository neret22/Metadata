# Import libraries, read metadata with astropy

```
import numpy as np
import matplotlib
import matplotlib.pyplot as plt

from astropy.io import fits
import matplotlib.pyplot as plt
from astropy.visualization import astropy_mpl_style
##read the metadata of fits image
plt.style.use(astropy_mpl_style)
hdulist=fits.open("1.fits")## check the metadata
hdulist.info()
print(repr(hdulist[0].header))## check the metadata
imagedata = hdulist[0].data
print(type(imagedata))
print(imagedata.shape)
```

# Plot fits image with "magma" color palette

```
plt.imshow(imagedata, cmap="magma")
plt.colorbar()
```

# Find min, max, mean and standard deviation counts values of image. Find the counts sum


```
print('Min:', np.min(imagedata))
print('Max:', np.max(imagedata))
print('Mean:', np.mean(imagedata))
print('Stdev:', np.std(imagedata))
```

```
np.sum(imagedata)
```
